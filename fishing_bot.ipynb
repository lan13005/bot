```python
def process_image (original_image):

    processed_image = cv2.cvtColor (original_image, cv2.COLOR_BGR2GRAY)
    
    processed_image = cv2.Canny (processed_image, threshold1 = 200, threshold2 = 300)
    return processed_image
```

```python
def getResid(goal,template):
    img = cv2.imread (goal)
    gray_frame = cv2.cvtColor (img, cv2.COLOR_RGB2GRAY)
    res = cv2.matchTemplate (gray_frame, template, cv2.TM_CCOEFF_NORMED)
    return res, gray_frame
```

```python
def getCentroid(res,template,gray_frame,makePlot=True):
    highIntensity=np.where(res>0.6,1,0)
    centroidX=highIntensity.sum(axis=0).argmax()
    centroidY=highIntensity.sum(axis=1).argmax()
    centroidX=centroidX+template.shape[1]/2
    centroidY=centroidY+template.shape[0]/2
    if makePlot:
        plt.imshow(gray_frame)
        plt.plot(centroidX,centroidY,"ro")
    return centroidX, centroidY
```
